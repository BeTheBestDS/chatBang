# 건설사 top10 컬럼 추가

- 1 ~ 5 : 삼성물산, 현대건설, 대우건설, 현대엔지니어링, 지에스건설
- 6 ~ 10 : 디엘이앤씨(대림산업), 포스코이앤씨(포스코건설), 롯데건설, 에스케이에코플랜트(에스케이건설), 호반건설

```python
df['top10'] = df['건설사_ex'].apply(lambda x: 1 if '삼성물산' in x
                                 else (1 if '현대건설' in x
                                       else (1 if '대우건설' in x
                                             else (1 if '현대엔지니어링' in x
                                                   else (1if 'GS건설' in x
                                                         else (1 if 'DL이앤씨' in x
                                                               else (1 if 'POSCO이앤씨' in x
                                                                     else (1 if '롯데건설' in x
                                                                           else (1 if 'SK에코플랜트' in x
                                                                                 else (1 if '호반건설' in x
                                             else 0))))))))))
```
