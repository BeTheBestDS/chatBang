```python
from selenium.webdriver.common.by import By
from selenium import webdriver
```

```python
import numpy as np
import pandas as pd
```

```python
school_url = 'https://asil.kr/asil/sub/school_list.jsp'
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-gpu')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver = webdriver.Chrome(options=options)
```

```python
driver.get(school_url)
```

```python
do_list = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/select[1]').text.split()
print(len(do_list))
str_columns = '도 순위 시군구 동읍면리 학교명 응시자수 평균 국어 영어 수학 특목고진학률 특목고진학수 과학고/외고국제고 졸업자수\n'

for num in range(1, len(do_list)) :
    driver.find_element(By.XPATH, f'/html/body/div/div[1]/div[1]/select[1]/option[{num+1}]').click()

    gu_list = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/select[2]').text.strip().split('\n')
    print(len(gu_list))
    for cnt in range(1, len(gu_list)) :
        driver.find_element(By.XPATH, f'/html/body/div/div[1]/div[1]/select[2]/option[{cnt+1}]').click()

        school = driver.find_element(By.XPATH, '/html/body/div/div[2]/table/tbody').text.split('\n')

        if len(school) > 1 :
            file_txt = open(f'/content/drive/MyDrive/Final/Crawling/school/{do_list[num]}_{gu_list[cnt]}.txt', 'w')
            file_txt.write(str_columns)
            for i in range(0, len(school), 2) :
                if len(school[i].split()) == 11 :
                    file_txt.write(do_list[num]+' '+school[i]+' '+school[i+1]+'\n')
                elif len(school[i].split()) == 12 :
                    school[i] = school[i].split()
                    school[i].pop(3)
                    school[i] = ' '.join(school[i])
                    file_txt.write(do_list[num]+' '+school[i]+' '+school[i+1]+'\n')
                else :
                    school[i] = school[i].split()
                    school[i].insert(2, '-')
                    school[i] = ' '.join(school[i])
                    file_txt.write(do_list[num]+' '+school[i]+' '+school[i+1]+'\n')
            file_txt.close()
driver.close()
```
