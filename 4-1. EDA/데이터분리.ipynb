```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
total_data = pd.read_csv('total_data.csv', encoding='cp949')
total_data.shape
```

아파트 단지 정보만 추출

```python
apt_data = total_data.drop_duplicates('도로명주소')
apt_data.shape
```

17년 3분기 부터, 결측치 연속으로 6분기 미만인 분기별 거래 데이터만 추출

```python
target_data = total_data[total_data['분기']>'2017 2Q']
target_data['temp'] = target_data['도로명주소'] + '_' + target_data['전용면적구간'].astype(str)
group = target_data.groupby(['temp'])['결측치연속'].agg(['max']).reset_index()
to_del = group[group['max']>5]['temp']
target_data = target_data[~target_data['temp'].isin(to_del)]
target_data.drop('temp', axis=1, inplace=True)
target_data.shape
```
